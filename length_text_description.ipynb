```python
import pandas  as pd 
import os 
interval = 10
interval_list = []
freq_list = []
```

```python
df = pd.read_csv('text_discrption_length.csv')
for index, row in df.iterrows():
    print(row['The length of text discprition'], row['Number of each length'])
    if row['The length of text discprition'] <= interval:
        if interval not in interval_list:
            interval_list.append(interval)
            freq_list.append(row['Number of each length'])
        else:
            idi = interval_list.index(interval)
            freq_list[idi] += row['Number of each length']
    else:
        while row['The length of text discprition'] > interval:
            interval += 10
        if interval not in interval_list:
            interval_list.append(interval)
            freq_list.append(row['Number of each length'])
        else:
            idi = interval_list.index(interval)
            freq_list[idi] += row['Number of each length']
```

```python
df = pd.DataFrame({'The length of text discription':interval_list,'Number of text discription':freq_list})
df.to_csv('text_disctiption_length_interval.csv',index = False)
```
